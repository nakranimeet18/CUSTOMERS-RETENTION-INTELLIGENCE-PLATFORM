```python
"""
==================================================================================
 CUSTOMER RETENTION INTELLIGENCE PLATFORM - SECONDARY DATASET (Olist)
 Phase 1: Data Understanding -> File 1 of 5: olist_products_dataset.csv
==================================================================================
 Purpose : Explore product catalogue data (used for Tableau dashboard).
           Contains physical attributes and category of every product sold.
==================================================================================
"""

import pandas as pd
import numpy as np

pd.set_option("display.max_columns", None)
pd.set_option("display.width", 150)


def section(title: str) :
    print("\n" + "=" * 90)
    print(f" {title}")
    print("=" * 90)


# ----------------------------------------------------------------------------------
# 1. DATA COLLECTION
# ----------------------------------------------------------------------------------
section("1. DATA COLLECTION")

DATA_PATH = "/Users/meetnakrani/Library/Mobile Documents/com~apple~CloudDocs/Customer-Retention-intelligence-Platform/DataSet/RAW/secondary Data/olist_products_dataset.csv"
df = pd.read_csv(DATA_PATH)

print(f"Dataset loaded successfully from: {DATA_PATH}")
print(f"Shape of dataset -> Rows: {df.shape[0]}, Columns: {df.shape[1]}")


# ----------------------------------------------------------------------------------
# 2. DATASET STRUCTURE
# ----------------------------------------------------------------------------------
section("2. DATASET STRUCTURE (First & Last Records)")
print("\n--- First 5 records ---")
print(df.head())
print("\n--- Last 5 records ---")
print(df.tail())

section("2.1 COLUMN NAMES & DATA TYPES")
print(df.dtypes)

section("2.2 MEMORY USAGE & GENERAL INFO")
df.info(memory_usage="deep")


# ----------------------------------------------------------------------------------
# 3. MISSING VALUE ANALYSIS
# ----------------------------------------------------------------------------------
section("3. MISSING VALUE ANALYSIS")

missing_count = df.isnull().sum()
missing_percent = (missing_count / len(df)) * 100
missing_summary = pd.DataFrame({
    "Missing Count": missing_count,
    "Missing %": missing_percent.round(2)
})
missing_summary = missing_summary[missing_summary["Missing Count"] > 0].sort_values("Missing %", ascending=False)

if missing_summary.empty:
    print("No missing values found in the dataset.")
else:
    print(missing_summary)


# ----------------------------------------------------------------------------------
# 4. DUPLICATE RECORD CHECK
# ----------------------------------------------------------------------------------
section("4. DUPLICATE RECORD CHECK")
print(f"Fully duplicated rows       : {df.duplicated().sum()}")
print(f"Duplicated product_id count : {df['product_id'].duplicated().sum()}")


# ----------------------------------------------------------------------------------
# 5. FEATURE SEGREGATION
# ----------------------------------------------------------------------------------
section("5. FEATURE SEGREGATION (Numerical vs Categorical)")

numerical_cols = df.select_dtypes(include=[np.number]).columns.tolist()
categorical_cols = df.select_dtypes(include=["object"]).columns.tolist()
if "product_id" in categorical_cols:
    categorical_cols.remove("product_id")   # identifier, not a dashboard dimension

print(f"Numerical columns   ({len(numerical_cols)}): {numerical_cols}")
print(f"Categorical columns ({len(categorical_cols)}): {categorical_cols}")


# ----------------------------------------------------------------------------------
# 6. STATISTICAL SUMMARY - NUMERICAL FEATURES (dimensions & weight/size)
# ----------------------------------------------------------------------------------
section("6. STATISTICAL SUMMARY - NUMERICAL FEATURES")
print(df[numerical_cols].describe().T)


# ----------------------------------------------------------------------------------
# 7. CATEGORY ANALYSIS (Product Category Name)
# ----------------------------------------------------------------------------------
section("7. PRODUCT CATEGORY ANALYSIS")

print(f"Total unique product categories: {df['product_category_name'].nunique()}")
print("\nTop 15 categories by product count:")
print(df["product_category_name"].value_counts().head(15))

print(f"\nProducts with missing category: {df['product_category_name'].isnull().sum()}")


# ----------------------------------------------------------------------------------
# 8. CARDINALITY CHECK
# ----------------------------------------------------------------------------------
section("8. CARDINALITY CHECK")
cardinality = df.nunique().sort_values()
print(cardinality)

id_like_cols = cardinality[cardinality == len(df)].index.tolist()
print(f"\nID-like columns (fully unique): {id_like_cols if id_like_cols else 'None'}")


# ----------------------------------------------------------------------------------
# 9. INITIAL DATA QUALITY OBSERVATIONS
# ----------------------------------------------------------------------------------
section("9. INITIAL DATA QUALITY OBSERVATIONS")

observations = f"""
1. Dataset contains {df.shape[0]} products described by {df.shape[1]} attributes.
2. 'product_id' is a unique identifier and acts as the join key to
   'olist_order_items_dataset.csv' for building the Tableau data model.
3. {missing_summary.shape[0]} columns contain missing values, mainly product
   description/name length and category name -> for Tableau, blanks should
   be relabeled (e.g. 'Unknown Category') rather than dropped, since dropping
   rows would break the join with order_items.
4. No duplicate product_id records were found.
5. There are {df['product_category_name'].nunique()} distinct product categories in
   Portuguese (source dataset) -> consider translating to English for the
   dashboard using Olist's official category translation file, if needed.
"""
print(observations)

section("PRODUCTS DATASET - DATA UNDERSTANDING COMPLETE")
```
