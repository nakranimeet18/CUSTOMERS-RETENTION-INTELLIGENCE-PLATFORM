Business Problem
My project work on an  e-commerce company.

Every month:

* Thousands of customers register.
* Some customers buy regularly.
* Some stop purchasing after a few weeks.
* The company doesn’t know who is likely to leave.

This creates problems such as:

* Loss of revenue.
* Increased marketing costs to acquire new customers.
* Reduced customer loyalty.
* Difficulty planning marketing campaigns.

The business wants to answer:

* Which customers are likely to leave?
* Why are they leaving?
* Which customers should we focus on first?
* What actions can reduce customer churn?

Project Goal

The main goal of my  project is:

To build a Customer Retention Intelligence Platform that predicts customers who are at risk of churning and provides insights that help businesses improve customer retention.

⸻

Business Objectives

Your project aims to:

* Reduce customer churn.
* Improve customer satisfaction.
* Increase repeat purchases.
* Identify high-risk customers early.
* Support data-driven business decisions.
* Provide an interactive analytics dashboard.

Business Questions

Before building the model, the business wants answers to questions like:

* Which customers are most likely to churn?
* Which factors contribute most to churn?
* Does customer satisfaction affect retention?
* Does purchase frequency affect churn?
* Do complaints increase the likelihood of churn?
* Which customer segments require retention campaigns?


Why is Data Science Needed?

Without data science:

* Businesses rely on intuition.
* High-risk customers may be identified too late.
* Marketing campaigns may target the wrong audience.

With data science:

* Customer behavior is analyzed systematically.
* Churn probability is predicted before customers leave.
* Retention efforts can be prioritized for customers who need them most.


Business Success Criteria

A successful project should:

* Predict churn accurately.
* Identify the key factors influencing churn.
* Present clear visual insights through dashboards.
* Allow new customer data to be entered through the Streamlit application.
* Support informed business decisions.
